```python
# Lista de frecuencias para asfreq():
#Por defecto, cuando usas el método asfreq('D') en pandas, si tienes varias entradas en el mismo día, solo se mantendrá la última entrada de ese día y las demás s
#Asfrec sirve para crear suma por valor dado

# 'D'   : Diario (día)
# 'B'   : Días hábiles (excluye fines de semana)
# 'W'   : Semanal (inicio o fin de semana)
# 'M'   : Fin de mes
# 'SM'  : Medio mes (1ro y 15 de cada mes)
# 'Q'   : Fin de trimestre
# 'A'   : Fin de año
# 'H'   : Horario (una entrada por hora)
# 'T'   : Minutal (una entrada por minuto)
# 'S'   : Segundos
# 'L'   : Milisegundos
# 'U'   : Microsegundos
# 'N'   : Nanosegundos

# Modificadores adicionales:
# 'W-MON' : Semanas que comienzan en lunes
# 'Q-JAN' : Trimestres que terminan en enero

df_comp = df_comp.asfreq('b')
```
